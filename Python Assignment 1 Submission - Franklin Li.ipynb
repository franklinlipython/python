```python
#Updated Submission

word_a = "aab" #Silent for the original example 
word_b = "ab" #Listen for the original example -> silent and listen are anagrams
 
def anagram_checker(word_a, word_b):
    letters_a = list(word_a.lower()) # aab becomes [a, a, b]
    letters_b = list(word_b.lower()) # ab becomes [a, b] 
    for a in letters_a: #for each letter in the first word
        if a in letters_b:
            letters_b.remove(a)
        else:
            return False
    if letters_b == []:
        return True
    else:
        return False            

anagram_checker(word_a, word_b)
```

```python
word_a = "Aab"
word_b = "ab"
is_case_sensitive = True

def anagram_checker(word_a, word_b, is_case_sensitive):
    
    if is_case_sensitive == False:
        letters_a = list(word_a.lower())  
        letters_b = list(word_b.lower())  
        for a in letters_a:
            if a in letters_b:
                letters_b.remove(a)
            else:
                return False
        if letters_b == []: 
            return True
        else:
            return False
        
    else: #if is_case_sensitive == True , THEN do not convert into lowercase letters but still convert to list
        letters_a_case_sensitive = list(word_a) # [A, a, b]
        letters_b_case_sensitive = list(word_b) # [a, b]
        for a_case_sensitive in letters_a_case_sensitive:
            if a_case_sensitive in letters_b_case_sensitive:
                letters_b_case_sensitive.remove(a_case_sensitive)
            else:
                return False
        if letters_b_case_sensitive == []:
            return True
        else:
            return False
        
anagram_checker(word_a, word_b, is_case_sensitive)
```
