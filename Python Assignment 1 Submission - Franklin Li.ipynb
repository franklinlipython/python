```python
#Updated Submission

word_a = "bye"
word_b = "Listen"
 
def anagram_checker(word_a, word_b):
    letters_a = list(word_a.lower()) # ['S', 'i', 'l', 'e', 'n', 't']
    letters_b = list(word_b.lower()) # ['L', 'i', 's', 't', 'e', 'n'] 
    for a in letters_a:
        while a in letters_b:
            letters_b.remove(a)
    if letters_b == []:
        return True
    else:
        return False            

anagram_checker(word_a, word_b)
```

```python
word_a = "siLent"
word_b = "LiSten"
is_case_sensitive = True

def anagram_checker(word_a, word_b, is_case_sensitive):
    
    if is_case_sensitive == False:
        letters_a = list(word_a.lower()) # ['S', 'i', 'l', 'e', 'n', 't']
        letters_b = list(word_b.lower()) # ['L', 'i', 's', 't', 'e', 'n']
        for a in letters_a:
            while a in letters_b:
                letters_b.remove(a)
        if letters_b == []: 
            return True
        else:
            return False
        
    else: #if is_case_sensitive == False , THEN do not convert into lowercase letters [Silent and listen]
        letters_a_case_sensitive = list(word_a) 
        letters_b_case_sensitive = list(word_b)
        for a_case_sensitive in letters_a_case_sensitive:
            while a_case_sensitive in letters_b_case_sensitive:
                letters_b_case_sensitive.remove(a_case_sensitive)
        if letters_b_case_sensitive == []:
            return True
        else:
            return False
        
anagram_checker(word_a, word_b, is_case_sensitive)
```
